```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter to include only numerical columns
numerical_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# Display the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of MEP Features')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'committeeCount' is a valid column, you might need to create or verify it
# For demonstration, let's assume it's a numerical column representing the number of committees an MEP is part of
# If it doesn't exist, you would need to create it based on other available data

# Select numerical features for clustering
features = df[['Number of past legislatures in the EP (not considering the 2019-2024)', 'committeeCount']] 

# Handle missing values and scale the features
features = features.fillna(0)
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # You can change the number of clusters
df['Cluster'] = kmeans.fit_predict(features_scaled)

# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(features_scaled[:, 0], features_scaled[:, 1], c=df['Cluster'], cmap='viridis', marker='o')
plt.title('KMeans Clustering of MEPs')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.colorbar()
plt.show()
```

```python
political_distribution = df.groupby(['country', 'politicalGroup']).size().unstack(fill_value=0)
print(political_distribution.head())
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(political_distribution, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Political Distribution of MEPs per Country and Party')
plt.xlabel('Political Group')
plt.ylabel('Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
